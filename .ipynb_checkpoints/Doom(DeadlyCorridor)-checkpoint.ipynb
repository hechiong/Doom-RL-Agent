```python
# Import vizdoom for game env
from vizdoom import *
# Import random for action sampling
import random
# Import time for sleeping
import time
# Import numpy for identity matrix
import numpy as np
```

```python
# Set up game
game = DoomGame()
game.load_config("github/VizDoom/scenarios/deadly_corridor.cfg")
game.init()
```

```python
# This is the set of actions that can be taken in the environment
actions = np.identity(7, dtype=np.uint8)
```

```python
# Loop through episodes
episodes = 1
for episode in range(episodes):
    # Create a new episode/game
    game.new_episode()
    # Check the game isn't done
    while not game.is_episode_finished():
        # Get the game state
        state = game.get_state()
        # Get the game image
        img = state.screen_buffer
        # Get the game variables - ammo
        info = state.game_variables
        # Take an action
        # Second argument "skips 4 frames", giving reward 4 frames later
        #   Need this because when shooting, bullet doesn't immediately hit mob
        reward = game.make_action(random.choice(actions), 4)
        # Print reward
        print("Reward: ", reward)
        time.sleep(0.02)
    print("Result: ", game.get_total_reward())
    time.sleep(2)
```

# 2. Converting it to a Gym Environment

```python
!pip install gym
```

```python
# Import environment base class from OpenAI Gym
from gym import Env
# Import gym spaces
from gym.spaces import Box, Discrete
# Import opencv
import cv2
# Import fileinput for making inline changes
import fileinput
```

```python
# Config file
config = "github/VizDoom/scenarios/deadly_corridor.cfg"

# Create ViZDoom OpenAI Gym Environment
class ViZDoomGym(Env):
    # Function that is called when we start the env
    def __init__(self, render=False, config=config, doomSkill=5):
        # Inherit from Env
        super().__init__()
        
        # Set up the game
        self.game = DoomGame()
        
        # Set the doom skill / difficulty of enemies
        self.config = config
        self.setDoomSkill(doomSkill)
        
        self.game.load_config(self.config)
        
        # Render frame logic
        self.game.set_window_visible(render)
        
        # Start the game
        self.game.init()
        
        # Create the action space and observation space
        self.action_space = Discrete(7)
        self.observation_space = Box(low=0, high=255, shape=(100,160,1), dtype=np.uint8)
        
        # Game variables: HEALTH DAMAGE_TAKEN KILLCOUNT SELECTED_WEAPON_AMMO
        self.damage_taken = 0
        self.killcount = 0
        self.ammo = -1

    # This is how to take a step in the environment
    def step(self, action):
        actions = np.identity(7, dtype=np.uint8)
        movement_reward = self.game.make_action(actions[action], 4)
        
        state = self.game.get_state()
        reward = 0
        
        # Get all the information needed to be returned
        if state:
            img = state.screen_buffer
            img = self.grayscale(img)
            
            # Reward shaping
            game_variables = state.game_variables
            health, damage_taken, killcount, ammo = game_variables
            
            # Calculate reward deltas
            damage_taken_delta = -damage_taken + self.damage_taken
            self.damage_taken = damage_taken
            killcount_delta = killcount - self.killcount
            self.killcount = killcount
            
            if self.ammo >= 0:
                ammo_delta = ammo - self.ammo
            else:
                ammo_delta = 0
            self.ammo = ammo
            
            reward = movement_reward + damage_taken_delta*13.75 + killcount_delta*1000 + ammo_delta*20
            #reward /= 1000
            info = {"health":health,
                    "damage_taken":damage_taken,
                    "killcount":killcount,
                    "ammo":ammo}
        else:
            img = np.zeros(self.observation_space.shape)
            info = {"health":0,
                    "damage_taken":100,
                    "killcount":0,
                    "ammo":0}

        done = self.game.is_episode_finished()
        
        return img, reward, done, info
    
    # Define how to render the game or environment
    def render(self):
        pass
    
    # Start a new game
    def reset(self):
        self.game.new_episode()
        img = self.game.get_state().screen_buffer
        return self.grayscale(img)
    
    # Grayscale the game frame and resize it
    def grayscale(self, observation):
        gray = cv2.cvtColor(np.moveaxis(observation, 0, -1), cv2.COLOR_BGR2GRAY)
        resize = cv2.resize(gray, (160,100), interpolation=cv2.INTER_CUBIC)
        img = np.reshape(resize, (100,160,1))
        return img

    # Sets the doom skill variable in config file
    def setDoomSkill(self, doomSkill):
        for line in fileinput.input(self.config, inplace=1):
            if line.startswith("doom_skill"):
                print("doom_skill = {}".format(doomSkill), end="\n")
            else:
                print(line, end="")
        fileinput.close()
    
    # Close down the game
    def close(self):
        self.game.close()
```

# 4. Setup Callback

```python
!pip install torch torchvision torchaudio
```

```python
# Import os for file nav
import os
# Import callback class for sb3
from stable_baselines3.common.callbacks import BaseCallback
```

```python
class TrainAndLoggingCallback(BaseCallback):
    
    def __init__(self, check_freq, save_path, verbose=1):
        super(TrainAndLoggingCallback, self).__init__(verbose)
        self.check_freq = check_freq
        self.save_path = save_path

    def _init_callback(self):
        if self.save_path is not None:
            os.makedirs(self.save_path, exist_ok=True)

    def _on_step(self):
        if self.n_calls % self.check_freq == 0:
            model_path = os.path.join(self.save_path, "best_model_{}".format(self.n_calls))
            self.model.save(model_path)

        return True
```

```python
CHECKPOINT_DIR = "./train/train_deadlycorridor"
LOG_DIR = "./logs/log_deadlycorridor"
```

```python
callback = TrainAndLoggingCallback(check_freq=10000, save_path=CHECKPOINT_DIR)
```

# 5. Train Our Model Using Curriculum

```python
# Import PPO for training
from stable_baselines3 import PPO
```

```python
# Difficulty level of enemies
difficulty = 1
```

```python
# Non-rendered environment
env = ViZDoomGym(doomSkill=difficulty)
```

```python
# Increase n_steps and decrease clip_range, and vice versa
# gae_lambda doesn't change much if above relationship is maintained
model = PPO("CnnPolicy", env, verbose=1, tensorboard_log=LOG_DIR, learning_rate=1e-5, n_steps=8192,
            clip_range=0.1, gamma=0.95, gae_lambda=0.9)
```

```python
save_path = os.path.join(CHECKPOINT_DIR, "Doom Skill {}".format(difficulty))
callback = TrainAndLoggingCallback(check_freq=10000, save_path=save_path)

model.set_env(env)
model.learn(total_timesteps=600000, callback=callback)
```

```python
# Applying curriculum learning in non-rendered environments
for doomSkill in range(1, 6):
    env = ViZDoomGym(doomSkill=doomSkill)
    save_path = os.path.join(CHECKPOINT_DIR, "Doom Skill {}".format(doomSkill))
    callback = TrainAndLoggingCallback(check_freq=10000, save_path=save_path)
    
    model.set_env(env)
    model.learn(total_timesteps=600000, callback=callback)
```

```python
env.close()
```

```python
num_episodes = 5
for episode in range(num_episodes):
    obs = env.reset()
    done = False
    total_reward = 0
    while not done:
        action, _ = model.predict(obs)
        obs, reward, done, info = env.step(action)
        time.sleep(0.1)
        total_reward += reward
    print("Total reward for episode {}: {}".format(episode, total_reward))
    time.sleep(1)
```
